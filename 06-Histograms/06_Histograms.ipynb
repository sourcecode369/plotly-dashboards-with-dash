```
%%time
import plotly.offline as pyo
import plotly.graph_objs as go
import pandas as pd
import numpy as np
```

```
!git clone https://github.com/Pierian-Data/Plotly-Dashboards-with-Dash.git
```

```
df = pd.read_csv('Plotly-Dashboards-with-Dash/Data/mpg.csv')
df.head()
```

```
data = [go.Histogram(x=df["mpg"], xbins=dict(start=0,end=50,size=10))]
layout = go.Layout(title="MPG Histogram Distribution")
fig = go.Figure(data=data, layout=layout)
pyo.iplot(fig)
```

```
df = pd.read_csv('Plotly-Dashboards-with-Dash/Data/abalone.csv')
data = [go.Histogram(x=df["length"], xbins=dict(start=0,end=1,size=0.02))]
layout = go.Layout(title="Length Histogram Distribution")
fig = go.Figure(data=data, layout=layout)
pyo.iplot(fig)
```

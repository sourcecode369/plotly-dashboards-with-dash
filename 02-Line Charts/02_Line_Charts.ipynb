```
import numpy as np
import plotly.offline as pyo
import plotly.graph_objs as go 
np.random.seed(42)
```

```
x_values = np.linspace(0,1,100)
y_values = np.random.randn(100)
```

```
trace0 = go.Scatter(x=x_values, y=y_values+5,
                    mode="markers",name="markers")

trace1 = go.Scatter(x=x_values,y=y_values,
                    mode="lines", name="lines")

trace2 = go.Scatter(x=x_values, y=y_values-5,
                    mode="lines+markers",name="combined")
```

```
data = [trace0, trace1, trace2]
```

```
layout = go.Layout(title="Line Charts")
```

```
fig = go.Figure(data=data, layout=layout)
```

```
pyo.iplot(fig)
```

```
import plotly.offline as pyo
import plotly.graph_objs as go
import numpy as np
import pandas as pd
```

```
df = pd.read_csv('Plotly-Dashboards-with-Dash/Data/2010YumaAZ.csv')
```

```
df.head()
```

```
days = list(df["DAY"].unique())
```

```
days
```

```
data = [go.Scatter(x = df[df["DAY"]==day]["LST_TIME"],y=df[df["DAY"]==day]["T_HR_AVG"], mode="markers+lines", name=day) for day in days]
layout = go.Layout(title="LineChart Excerise", xaxis=dict(title="LST_TIME"), yaxis=dict(title="T_HR_AVG") ,hovermode="closest")
```

```
fig = go.Figure(data=data, layout=layout)
pyo.iplot(fig)
```

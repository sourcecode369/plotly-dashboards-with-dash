```
%%time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import plotly.offline as pyo
import plotly.figure_factory as ff
import plotly.graph_objs as go
```

```
x1 = np.random.randn(200)

hist_data = [x1]
group_labels = ['X1']
fig = ff.create_distplot(hist_data=hist_data,group_labels=group_labels)
pyo.iplot(fig)
```

```
x1 = np.random.randn(200)-2
x2 = np.random.randn(200)+2
x3 = np.random.randn(200)+4
x4 = np.random.randn(200)

hist_data = [x1,x2,x3,x4]
group_labels = ['X1',"X2","X3","X4"]
fig = ff.create_distplot(hist_data=hist_data,group_labels=group_labels,bin_size=[.2,.1,.4,.3])
pyo.iplot(fig)
```

```
df = pd.read_csv('Plotly-Dashboards-with-Dash/Data/iris.csv')
```

```
df.head()
```

```
classes = df['class'].unique()
```

```
trace1 = df[df["class"]==classes[0]]['petal_length']
trace2 = df[df["class"]==classes[1]]['petal_length']
trace3 = df[df["class"]==classes[2]]['petal_length']
```

```
data = [trace1,trace2,trace3]
labels = [classes[0],classes[1],classes[2]]
figure = ff.create_distplot(hist_data=data,group_labels=labels)
pyo.iplot(figure)
```
